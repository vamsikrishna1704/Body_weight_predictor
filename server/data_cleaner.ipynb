```python
import pandas as pd
```

```python
data = pd.read_csv("./data/Dataset.csv")
```

```python
data.head(10)
```

```python
data = data.drop(columns=["Unnamed: 0","BMI_tags", "Label"])
data
```

```python
data
```

```python
def convert_height_to_cms(height):
    return height * 100

data["height(m)"] = data["height(m)"].apply(convert_height_to_cms)
data
```

```python
data = data.loc[data["age"]>=13]
```

```python
data_men = data.loc[data["gender"] == "M"]
data_female = data.loc[data["gender"] == "F"]

data_men["Cal/Day"] = (10 * data["weight(kg)"] + 6.25 * data["height(m)"] - 5 * data["age"] + 5) * 1.2
data_female["Cal/Day"] = (10 * data["weight(kg)"] + 6.25 * data["height(m)"] - 5 * data["age"] - 161) * 1.2
```

```python
data = pd.concat([data_men, data_female])
```

```python
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
data
```

```python
data = data.drop(columns=["activity_level", "calories_to_maintain_weight"])
```

```python
data.describe()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
data = pd.get_dummies(data, columns=["gender"])
data
```

```python
data.columns = ["age", "weight", "height", "bmi", "bmr", "Cal/Day", "is_female", "is_male"]
```
